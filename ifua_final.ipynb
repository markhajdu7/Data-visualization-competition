```python
#Data viz

plt.style.use('Solarize_Light2')
matplotlib.rcParams['figure.figsize'] = (5,4)

#sns.set_style('whitegrid')

image = sns.scatterplot(x ='Tiszta karbonmentes energia', y ='1kWh energiára jutó CO2 kibocsátás', 
            data = df, hue='Év', palette=['gold','green'], alpha=0.6,x_jitter=True,
    y_jitter=True,s=100
                    )

#plt.title('A széndioxid intenzitás és a tiszta karbonmentes energia arány alakulása országos bontásban',fontsize=11,color='k')
#plt.suptitle('Pozitív változások Európában 2012-2021 között a villamosenergia termelésben',fontsize=12, y=1,color='k')
plt.suptitle('A széndioxid intenzitás és a tiszta karbonmentes energia arány alakulása országos bontásban',fontsize=12, y=1,color='k')


plt.xlabel('Tiszta karbonmentes energia (%-ban, 100%= Összes termelt villamosenergia)',fontsize=10,color='k')
plt.ylabel('Termelt energia széndioxid intentizása (gCO2/kWh)',fontsize=10,color='k')

legend = plt.legend()
frame = legend.get_frame()
frame.set_facecolor('xkcd:eggshell')

# x-axis and y-axis values for plotting
x = [0, 20, 40, 60, 80, 100]
y = [0, 200, 400, 600, 785]
 
# labels for x-asix
labels = ['0', '20', '40', '60', '80', '100']
labelsy = ['0', '200', '400', '600', '800']
 
# setting x-axis values
plt.xticks(x, labels)

plt.yticks(y, labelsy)

plt.tight_layout
```
